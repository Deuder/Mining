```python
def labelfilter(df,cats,de):
    count = {}
    terms = {}
    #for all terms and categorys

    for j in range(len(cats)):
        #synonyms for viwed category
        cat = cats.iloc[j,1].lower()
        syns = cats.iloc[j,2].lower().split(", ")
        cnt = 0

        #view all terms of dataframe
        for i in range (len(df)):
            termcount = df.iloc[i,0]
            term = df.iloc[i,1].lower()

            #if term in synonyms for category
            if (term in syns):
                cnt += termcount

        if(cnt > 0):
            if(cat in terms):
                count[cat] += cnt
            else:
                count[cat] = cnt
                terms[cat]= cat  
        
    #convert to series and dataframe, sort by count column descending
    count = pd.Series(count)
    terms = pd.Series(terms)
    countdf = pd.DataFrame({'count':count, 'term':terms})
    countdf = countdf.sort_values(by = 'count', ascending = False)
    cats["count"] = 0
    
    for i in range (len(countdf)):
        count = countdf.iloc[i,0]
        term = countdf.iloc[i,1]
    
        cats.loc[cats['term'] == term , 'count'] = count
    
    cats = cats.drop('syns', axis = 1)
    cats = cats.drop('depth', axis = 1)
    
    for i in range (len(cats)):
        cat = cats.iloc[i,0]
        count = cats.iloc[i,2]
        for j in range (i+1,len(cats)): 
            cat2 = cats.iloc[j,1]
            if cat == cat2:
                cats.iloc[j,2]+= count
        
    cats = cats.drop('cat', axis = 1)
    if(de):
        cats.to_csv(r"C:\Users\thegy\Documents\Studium\HS FL 4\Mining\DFs\manualcategorycountde.csv", sep=";",decimal=",", index=False)
    else:
        cats.to_csv(r"C:\Users\thegy\Documents\Studium\HS FL 4\Mining\DFs\manualcategorycounten.csv", sep=";",decimal=",", index=True)
    print(cats)
```
