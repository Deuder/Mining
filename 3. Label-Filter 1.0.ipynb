```python
cats = r"C:\Users\thegy\Documents\Studium\HS FL 4\Mining\DFs\labeledclusterslow.csv"
cats = pd.read_csv(cats, delimiter = ";")
#cats: (index), cat, term, syns
i= 5
print(cats.iloc[i,0])
print(cats.iloc[i,1])
print(cats.iloc[i,2])
print(cats.iloc[i,3])
```

```python
count = {}
terms = {}
#for all terms and categorys

for j in range(len(cats)):
    #synonyms for viwed category
    cat = cats.iloc[j,1].lower()
    syns = cats.iloc[j,2].lower().split(", ")
    cnt = 0
    
    #view all terms of dataframe
    for i in range (len(df)):
        termcount = df.iloc[i,0]
        term = df.iloc[i,1].lower()
        
        #if term in synonyms for category
        if (term in syns):
            cnt += termcount
            
    if(cnt > 0):
        if(cat in terms):
            count[cat] += cnt
        else:
            count[cat] = cnt
            terms[cat]= cat  
        
```

```python
 #convert to series and dataframe, sort by count column descending
count = pd.Series(count)
terms = pd.Series(terms)
countdf = pd.DataFrame({'count':count, 'term':terms})
countdf = countdf.sort_values(by = 'count', ascending = False)
```

```python
cats["count"] = 0
```

```python
for i in range (len(countdf)):
    count = countdf.iloc[i,0]
    term = countdf.iloc[i,1]
    
    cats.loc[cats['term'] == term , 'count'] = count
```
